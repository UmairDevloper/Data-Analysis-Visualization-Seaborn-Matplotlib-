# Distribution Plots

These plots are used to visualize the Quantitative data in form of plots. It mainly done in Seaborn.
The important plots are :
1. distplot/histplot
2. jointplot
3. pairplot
4. rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading dataset using seaboen dataset

```python
df = sns.load_dataset("tips")
df
```

To check whether the given column is qualitative or quantitative. 

```python
df["size"].unique()
```

To check the distribution of data around the mean of a group

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"], kde=True)
```

To compare the two distributions together and find the trend

```python
sns.jointplot(x = "total_bill", y = "tip", data=df, kind="scatter")
sns.jointplot(x = "total_bill", y = "tip", data=df, kind="reg")
```

```python
sns.rugplot(df['tip'])
```

```python
sns.pairplot(df, hue='size', palette='rainbow', kind="reg")
```
