```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## lmplot()
regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

```python
tips = sns.load_dataset('tips')
tips
```

It tells that what we have to predict through which entity. 
So, X = through which we have to predict, Y = what to predict.
In below example we are predicting the tip through total bill

```python
sns.lmplot( x= 'total_bill', y = 'tip', data=tips)
```

```python
sns.lmplot( x= 'total_bill', y = 'tip', data=tips, hue='sex')
```

```python
sns.lmplot( x= 'total_bill', y = 'tip', data=tips,hue='sex', palette='rainbow')
```

```python
sns.lmplot( x= 'total_bill', y = 'tip', data=tips,hue='sex', palette='rainbow', markers=['o', 'v'], scatter_kws={'s':500})  # Changing the size
```
